## Configurção do ambiente

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = 'plotly_dark'
```

## Importar e selecionar os dados a serem trabalhados

```python
df = pd.read_csv('ocorrencia_2010_2020.csv', sep=';', parse_dates=['ocorrencia_dia'])
```

```python
df = df[['codigo_ocorrencia','codigo_ocorrencia2','ocorrencia_classificacao','ocorrencia_cidade','ocorrencia_uf',
        'ocorrencia_aerodromo', 'ocorrencia_dia', 'ocorrencia_hora','total_recomendacoes']]
df.rename({'ocorrencia_classificacao':'classificacao','ocorrencia_cidade':'cidade','ocorrencia_uf':'uf',
        'ocorrencia_aerodromo':'aerodromo', 'ocorrencia_dia':'dia', 'ocorrencia_hora':'hora'},axis=1, inplace=True)
```
